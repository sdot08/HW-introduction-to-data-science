$$\text{Mutual Information} = \sum_{y \in Y} \sum_{x \in X} p(x, y) \cdot log\frac{p(x, y)}{p(x)p(y)}$$

$$\text{Entropy} = H(Y) = -\sum_{y \in Y} p(y) \cdot log(p(y))$$

$$\text{Conditional Entropy} = H(Y \mid X) = \sum_{x \in X} p(x) \cdot H(Y \mid X = x)$$

1\. Load the data and prepare it for modeling. Note that the features are already processed for you, so the only thing needed here is split the data into training and testing.
````python
import numpy as np
import pandas as pd
colnames=['revenue','outcalls','incalls','months','eqpdays','webcaps','marryyes','travel','pcown','creditcd','retcalls','churndep']
data=pd.read_csv('cell2cell_data.csv',header=None, names=colnames)
train_pct=0.8
divider=np.random.random(data.shape[0])<train_pct
train_df=pd.DataFrame(data[divider])
test_df=pd.DataFrame(data[divider==False])
````

### Part 3 - Finding a Good Decision Tree
The default options for your decision tree may not be optimal. We need to analyze whether tuning the parameters can improve the accuracy of the classifier.  For the following options `min_samples_split` and `min_samples_leaf`:

1\. Generate a list of 10 values of each for the parameters mim_samples_split and min_samples_leaf. 

````python
min_samples_split_values = np.linspace(4,44,10).astype(int)
min_samples_leaf_values = np.linspace(2,42,10).astype(int)
````

2\. Explain in words your reasoning for choosing the above ranges.

The min_samples_split is the minimum number of samples required to split an internal node, and the min_samples_leaf is the minimum number of samples required to be at a leaf node. The default values of these two values are 2 and 1. If we want to avoid overfitiing, we should shoose the values that are bigger than the default values to compare the performance. I choose 44 and 42 because we need a relatively wide range of value to choose to obtain the best accuracy results. Another reason is that the accuracy performance start to drop when min_samples_split_values is bigger than 34. Therefore the optimal value should be in the range I shoose.

3\. For each combination of values in 3.1 (there should be 100), build a new classifier and check the classifier's accuracy on the test data. Plot the test set accuracy for these options. Use the values of `min_samples_split` as the x-axis and generate a new series (line) for each of `min_samples_leaf`.

````python
def plotAccTree(splits,accuracy,leaf):
    c = (np.random.rand(), np.random.rand(), np.random.rand())
    plt.plot(splits,accuracy,color=c,label='min_samples_leaf_values=%0.1f)'%leaf)
    plt.xlabel('min_samples_split_values')
    plt.ylabel('Accuracy')
    plt.title('Accuracy for Decision Trees of Different Parameters')
    plt.legend(loc="lower right")
````

````python
plt.figure()
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (12, 6))
ax = plt.subplot(111)

for leaf in min_samples_leaf_values:
    accuracy=[]
    splits=[]
    for split in min_samples_split_values:
        dtl= DecisionTreeClassifier(criterion='entropy',min_samples_split=split,min_samples_leaf=leaf)
        dtl.fit(X_train,Y_train )
        pred=dtl.predict_proba(X_test)[:,1]
        accuracy.append(1-np.average(np.abs(Y_test['churndep']-pred)))
        splits.append(split)
    plotAccTree(splits,accuracy,leaf)

box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.0 , box.width, box.height * 1])
ax.legend(loc = 'upper center', bbox_to_anchor = (0.5, -0.15), fancybox = True, 
              shadow = True, ncol = 4, prop = {'size':10})
plt.show()
````
